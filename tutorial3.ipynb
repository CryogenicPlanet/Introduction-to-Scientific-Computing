```python
def sine(x,n):
    sign = 1
    sum = 0.0
    for i in range(0,n):
        sum = sum + sign*((x**(2*i+1))/(np.math.factorial(2*i+1)))
        sign = sign*(-1)
    return sum
print(sine(0.1,20))
```
